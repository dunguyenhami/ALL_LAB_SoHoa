```python
import pandas as pd
import numpy as np

data = {
    "ID" : [ 101,102,103,104,105],
    "Name": ["Quynh Nhu","Hoang Hai",None,"Phuong Tuan", "Thien An"],
    "Age": [26,18,20,None,19],
    "Salary": [5000,35000,65000,100000,None]
    
}
df = pd.DataFrame(data)
print(df)
# điền giá trị thiếu trong cột Age = giá trị trung bình
df["Age"].fillna(df["Age"].mean(), inplace=True)
# Loại bỏ các dòng chứa dữ liệu thiếu
df.dropna(inplace=True)
# điền giá trị thiếu trong name = "Unknown" ArithneticError
df["Name"].fillna("Unknown", inplace=True)
# Điền giá trị thiếu trong Salary = phương pháp Interpolation 
df["Salary"].interpolate(methode = "linear",inplace=True) 
print(df)
```

```python
# cho data 
df1 = pd.DataFrame({
    "Id": [1,2,3],
"Score_A": [70,90,85]
})
df2= pd.DataFrame({
    "Id": [3,4,5],
    "Score_B": [62,91,75]
})
# Thực hiện Merge trên cột Id (Inner join, left join, outer join)
inner_merge = pd.merge(df1, df2, on="Id", how="inner")  # Inner Join
left_merge = pd.merge(df1, df2, on="Id", how="left")    # Left Join
outer_merge = pd.merge(df1, df2, on="Id", how="outer")  # Outer Join
# Nói DataFrame theo chiều dọc
vertical_concat = pd.concat([df1, df2], axis=00)
# Gộp df1 và df2 để điền giá trị thiếu
merged_fillna = pd.merge(df1, df2, on="Id", how="outer").fillna(0)
df1.set_index("Id").combine_first(df2.set_index("Id")).reset_index()

print("Inner Merge:\n", inner_merge)
print("Left Merge:\n", left_merge)
print("Outer Merge:\n", outer_merge)
print("Vertical Concatenation:\n", vertical_concat)
print("Merged and Filled Data:\n", merged_fillna)
```

```python
import pandas as pd
import numpy as np

# Tạo DataFrame với 100,000 hàng
data = pd.DataFrame({
    "ID": range(1, 100001),
    "Value": np.random.randint(1, 100, 100000)  # Đúng số lượng phần tử
})

# Tối ưu hóa bộ nhớ bằng cách chuyển kiểu dữ liệu
data["ID"] = data["ID"].astype("uint32")
data["Value"] = data["Value"].astype("uint8")

# Tìm 5 giá trị phổ biến nhất trong cột "Value"
top_values = data["Value"].value_counts().head(5)
print("🔹 5 giá trị phổ biến nhất trong cột 'Value':")
print(top_values)

# Sử dụng query để lọc nhanh hơn (tương đương với df[df["Value"] > 90])
filtered_data = data.query("Value > 90")
print("\n🔹 Số hàng có Value > 90:", len(filtered_data))
```

```python
# Biểu đồ nhiều đường

# Cho dữ liệu sau
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0,10,100)
y1 = np.sin(t)
y2 = np.cos(t)
y3 = np.sin(t) * np.cos(t)

# tạo biểu đồ các đường theo thời gian
plt.figure(figsize=(10, 6))
plt.plot(t, y1, label="sin(t)", linewidth=2)
plt.plot(t, y2, label="cos(t)", linewidth=2)
plt.plot(t, y3, label="sin(t) * cos(t)", linewidth=2)

plt.title("Biểu đồ nhiều đường theo thời gian")
plt.xlabel("Thời gian (t)")
plt.ylabel("Giá trị hàm số")
plt.legend()
plt.grid()

# Hiển thị biểu đồ
plt.show()
```

```python
# Biểu đồ thanh nhóm

#cho dữ liệu sau 
labels = ["Q1","Q2","Q3","Q4"]
A = [500,700,800,600]
B= [450,350,650,750]
C= [500,250,850,600]
# vẽ biểu đồ thanh nhóm thể hiện doanh thu A,B,C trong 4 quýquý
x = np.arange(len(labels))  # Vị trí của các nhãn trên trục x
width = 0.25  # Độ rộng của mỗi cột

# Vẽ biểu đồ thanh nhóm
plt.figure(figsize=(8,6))
plt.bar(x - width, A, width=width, label="Công ty A", color="#ff9999")
plt.bar(x, B, width=width, label="Công ty B", color="#66b3ff")
plt.bar(x + width, C, width=width, label="Công ty C", color="#99ff99")

# Cài đặt trục x
plt.xticks(ticks=x, labels=labels)

# Thêm tiêu đề và nhãn trục
plt.xlabel("Quý")
plt.ylabel("Doanh thu (triệu)")
plt.title("Doanh thu của các công ty trong 4 quý")

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

```python
# Biểu đồ tròn
Cty = ["A","B","C","D"]
thiphan= [30,25,25,20]
color = ["#ff9999","#66b3ff","#99ff99","#ffcc99"]

# Tạo biểu đồ tròn thể hiện tỉ lệ thị phần của 4 công ty 

import matplotlib.pyplot as plt

# Dữ liệu
Cty = ["A", "B", "C", "D"]
thiphan = [30, 25, 25, 20]
color = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]

# Vẽ biểu đồ tròn
plt.figure(figsize=(6,6))
plt.pie(thiphan, labels=Cty, autopct="%1.1f%%", colors=color, startangle=90, wedgeprops={'edgecolor': 'black'})

# Tiêu đề
plt.title("Thị phần của các công ty")

# Hiển thị biểu đồ
plt.show()
```

```python
# Biểu đồ phân tán 
#cho dữ liệu sau
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100)*300
colors = np.random.rand(100)

# tạo biểu đồ phân tán của hai biến ngẫu nhiên 
# Vẽ biểu đồ phân tán
plt.figure(figsize=(8,6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, cmap='viridis', edgecolors="black")

# Tiêu đề và nhãn trục
plt.title("Biểu đồ phân tán của hai biến ngẫu nhiên")
plt.xlabel("X values")
plt.ylabel("Y values")

# Thanh màu thể hiện giá trị
plt.colorbar(label="Màu sắc tương ứng")

# Hiển thị biểu đồ
plt.show()
```
