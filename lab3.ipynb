```python
import pandas as pd
import numpy as np

data = {
    "ID" : [ 101,102,103,104,105],
    "Name": ["Quynh Nhu","Hoang Hai",None,"Phuong Tuan", "Thien An"],
    "Age": [26,18,20,None,19],
    "Salary": [5000,35000,65000,100000,None]
    
}
df = pd.DataFrame(data)
print(df)
# Ä‘iá»n giÃ¡ trá»‹ thiáº¿u trong cá»™t Age = giÃ¡ trá»‹ trung bÃ¬nh
df["Age"].fillna(df["Age"].mean(), inplace=True)
# Loáº¡i bá» cÃ¡c dÃ²ng chá»©a dá»¯ liá»‡u thiáº¿u
df.dropna(inplace=True)
# Ä‘iá»n giÃ¡ trá»‹ thiáº¿u trong name = "Unknown" ArithneticError
df["Name"].fillna("Unknown", inplace=True)
# Äiá»n giÃ¡ trá»‹ thiáº¿u trong Salary = phÆ°Æ¡ng phÃ¡p Interpolation 
df["Salary"].interpolate(methode = "linear",inplace=True) 
print(df)
```

```python
# cho data 
df1 = pd.DataFrame({
    "Id": [1,2,3],
"Score_A": [70,90,85]
})
df2= pd.DataFrame({
    "Id": [3,4,5],
    "Score_B": [62,91,75]
})
# Thá»±c hiá»‡n Merge trÃªn cá»™t Id (Inner join, left join, outer join)
inner_merge = pd.merge(df1, df2, on="Id", how="inner")  # Inner Join
left_merge = pd.merge(df1, df2, on="Id", how="left")    # Left Join
outer_merge = pd.merge(df1, df2, on="Id", how="outer")  # Outer Join
# NÃ³i DataFrame theo chiá»u dá»c
vertical_concat = pd.concat([df1, df2], axis=00)
# Gá»™p df1 vÃ  df2 Ä‘á»ƒ Ä‘iá»n giÃ¡ trá»‹ thiáº¿u
merged_fillna = pd.merge(df1, df2, on="Id", how="outer").fillna(0)
df1.set_index("Id").combine_first(df2.set_index("Id")).reset_index()

print("Inner Merge:\n", inner_merge)
print("Left Merge:\n", left_merge)
print("Outer Merge:\n", outer_merge)
print("Vertical Concatenation:\n", vertical_concat)
print("Merged and Filled Data:\n", merged_fillna)
```

```python
import pandas as pd
import numpy as np

# Táº¡o DataFrame vá»›i 100,000 hÃ ng
data = pd.DataFrame({
    "ID": range(1, 100001),
    "Value": np.random.randint(1, 100, 100000)  # ÄÃºng sá»‘ lÆ°á»£ng pháº§n tá»­
})

# Tá»‘i Æ°u hÃ³a bá»™ nhá»› báº±ng cÃ¡ch chuyá»ƒn kiá»ƒu dá»¯ liá»‡u
data["ID"] = data["ID"].astype("uint32")
data["Value"] = data["Value"].astype("uint8")

# TÃ¬m 5 giÃ¡ trá»‹ phá»• biáº¿n nháº¥t trong cá»™t "Value"
top_values = data["Value"].value_counts().head(5)
print("ðŸ”¹ 5 giÃ¡ trá»‹ phá»• biáº¿n nháº¥t trong cá»™t 'Value':")
print(top_values)

# Sá»­ dá»¥ng query Ä‘á»ƒ lá»c nhanh hÆ¡n (tÆ°Æ¡ng Ä‘Æ°Æ¡ng vá»›i df[df["Value"] > 90])
filtered_data = data.query("Value > 90")
print("\nðŸ”¹ Sá»‘ hÃ ng cÃ³ Value > 90:", len(filtered_data))
```

```python
# Biá»ƒu Ä‘á»“ nhiá»u Ä‘Æ°á»ng

# Cho dá»¯ liá»‡u sau
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0,10,100)
y1 = np.sin(t)
y2 = np.cos(t)
y3 = np.sin(t) * np.cos(t)

# táº¡o biá»ƒu Ä‘á»“ cÃ¡c Ä‘Æ°á»ng theo thá»i gian
plt.figure(figsize=(10, 6))
plt.plot(t, y1, label="sin(t)", linewidth=2)
plt.plot(t, y2, label="cos(t)", linewidth=2)
plt.plot(t, y3, label="sin(t) * cos(t)", linewidth=2)

plt.title("Biá»ƒu Ä‘á»“ nhiá»u Ä‘Æ°á»ng theo thá»i gian")
plt.xlabel("Thá»i gian (t)")
plt.ylabel("GiÃ¡ trá»‹ hÃ m sá»‘")
plt.legend()
plt.grid()

# Hiá»ƒn thá»‹ biá»ƒu Ä‘á»“
plt.show()
```

```python
# Biá»ƒu Ä‘á»“ thanh nhÃ³m

#cho dá»¯ liá»‡u sau 
labels = ["Q1","Q2","Q3","Q4"]
A = [500,700,800,600]
B= [450,350,650,750]
C= [500,250,850,600]
# váº½ biá»ƒu Ä‘á»“ thanh nhÃ³m thá»ƒ hiá»‡n doanh thu A,B,C trong 4 quÃ½quÃ½
x = np.arange(len(labels))  # Vá»‹ trÃ­ cá»§a cÃ¡c nhÃ£n trÃªn trá»¥c x
width = 0.25  # Äá»™ rá»™ng cá»§a má»—i cá»™t

# Váº½ biá»ƒu Ä‘á»“ thanh nhÃ³m
plt.figure(figsize=(8,6))
plt.bar(x - width, A, width=width, label="CÃ´ng ty A", color="#ff9999")
plt.bar(x, B, width=width, label="CÃ´ng ty B", color="#66b3ff")
plt.bar(x + width, C, width=width, label="CÃ´ng ty C", color="#99ff99")

# CÃ i Ä‘áº·t trá»¥c x
plt.xticks(ticks=x, labels=labels)

# ThÃªm tiÃªu Ä‘á» vÃ  nhÃ£n trá»¥c
plt.xlabel("QuÃ½")
plt.ylabel("Doanh thu (triá»‡u)")
plt.title("Doanh thu cá»§a cÃ¡c cÃ´ng ty trong 4 quÃ½")

# Hiá»ƒn thá»‹ chÃº thÃ­ch
plt.legend()

# Hiá»ƒn thá»‹ biá»ƒu Ä‘á»“
plt.show()
```
